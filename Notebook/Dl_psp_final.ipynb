```python
def removeNulls(dataframe, percent,corr_limit):
    dff = dataframe.copy()
    ishape = dff.shape
    colnames = (dff.isnull().sum()/len(dff))
    
    colnames = list(colnames[colnames.values>=percent].index)
    #print(colnames)
    rem_cols =[]
    for col in colnames:
        if np.absolute(corr[col])<=corr_limit:
            rem_cols.append(col)
    #print(rem_cols)
    dff.drop(labels = rem_cols,axis =1,inplace=True)        
    print("Number of Columns dropped: ",len(rem_cols))
    print("Old dataset columns",ishape,"New dataset columns",dff.shape)
    return dff
```

```python
from sklearn.metrics import mean_squared_error,accuracy_score
def eval_model(temp_model):
    bce,mse,acc = temp_model.evaluate(X_train, y_train, verbose=0)
    tbce,tmse,tacc = temp_model.evaluate(X_test, y_test, verbose=0)
    
    print("====== Training Score ======")
    print("Accuracy -> ",(acc*100))
    print("BCE -> ",(bce*100))
    print("MSE -> ",(mse*100))
    print("\n\n====== Testing Score ======")
    print("Accuracy -> ",(tacc*100))
    print("BCE -> ",(tbce*100))
    print("MSE -> ",(tmse*100))

eval_model(b_model)
```

### Fine-tuning Model

```python
import kerastuner as kt
from kerastuner import HyperModel
```

```python


def model_builder(hp):
       
       # Initialize the Sequential API and start stacking the layers
       model_tune=[]
       model_tune = keras.Sequential()
       model_tune.add(keras.layers.Flatten(input_shape=(X_train.shape[1],)))
       # Tune the number of units in the first Dense layer
       # Choose an optimal value between 32-512
       #hp1_units = hp.Int('units', min_value=25, max_value=35, step=1)
       model_tune.add(keras.layers.Dense(units=30, activation='relu'))
       # Add next layers
       hp2_units = hp.Int('units', min_value=4, max_value=10, step=1)
       model_tune.add(keras.layers.Dense(units=hp2_units, activation='relu'))
       hp_do = hp.Float('dropout', min_value=0.0, max_value=0.1, default=0.005, step=0.01)
       model_tune.add(layers.Dropout(rate=hp_do))
       model_tune.add(keras.layers.Dense(1,activation='sigmoid'))
       # Tune the learning rate for the optimizer
       # Choose an optimal value from 0.01, 0.001, or 0.0001
       #hp_learning_rate = hp.Choice('learning_rate', values=[0.1,0.01,0.001])
       #hp_optim = hp.Choice('optimizer', values=['SGD', 'RMSprop', 'Adagrad', 'Adadelta', 'Adam', 'Adamax', 'Nadam'])
       model_tune.compile(optimizer='rmsprop', loss='binary_crossentropy',metrics=['mse', 'accuracy'])
       return model_tune
```

```python

tuner_01= kt.Hyperband(model_builder, max_epochs=10, objective='mse', factor=3, 
    seed=42, executions_per_trial=2, directory='dirf')

```

```python
tuner_01.search_space_summary(extended=True) 
```

```python
stop_early = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)
#lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda x: 1. / (1. + x))
# Perform hypertuning
tuner_01.search(X_train, y_train, epochs=20, validation_split=0.3, callbacks=[stop_early] )
```

```python
best_hp=tuner_01.get_best_hyperparameters()[0]
best_hp.values
```

```python
# Build the model with the optimal hyperparameters
hy_model=[]
hy_model = tuner_01.hypermodel.build(best_hp)
hy_model.summary()
```

```python
hy_model.fit(X_train, y_train, epochs=10, validation_split=0.3)
```

```python
eval_model(hy_model)
```

### Fine-tuning Optimizers

```python
def model_builder_withOptim(hp):
    model_fine_tuner=[]
    hp_act = hp.Choice('activation', values=['softmax', 'softplus', 'softsign', 'relu', 'tanh', 'sigmoid', 'hard_sigmoid', 'linear'])
    model_fine_tuner = keras.Sequential()
    model_fine_tuner.add(keras.layers.Flatten(input_shape=(X_train.shape[1],)))
    model_fine_tuner.add(keras.layers.Dense(units=30, kernel_initializer='uniform', activation=hp_act))
    # Add next layers
    model_fine_tuner.add(keras.layers.Dense(units=10, kernel_initializer='uniform', activation=hp_act))
    model_fine_tuner.add(keras.layers.Dense(1,activation='sigmoid'))
    hp_optim = hp.Choice('optimizer', values=['SGD', 'RMSprop', 'Adagrad', 'Adadelta', 'Adam', 'Adamax', 'Nadam'])
    model_fine_tuner.compile(optimizer=hp_optim, loss='binary_crossentropy',metrics=['mse', 'accuracy'])
    return model_fine_tuner
```

```python
opt_tuner = kt.Hyperband(model_builder_withOptim,objective='val_accuracy', max_epochs=20,
    factor=2, directory='dir', project_name='optim_hyper')
```

```python
stop_early = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=4)
opt_tuner.search(X_train, y_train, epochs=20, validation_split=0.25)
```

```python
opt_tuner.results_summary()
```

```python
best_hps=opt_tuner.get_best_hyperparameters()[0]
best_hps.values
```

```python
hyper_model=[]
hyper_model = opt_tuner.hypermodel.build(best_hps)
hyper_model.summary()
```

```python
hyper_model.fit(X_train, y_train, epochs=20, validation_split=0.2)
```

```python
eval_model(hyper_model)
```
